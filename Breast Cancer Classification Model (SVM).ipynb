```python
df_cancer = pd.DataFrame(np.c_[cancer['data'],cancer['target']], columns = np.append(cancer['feature_names'],['target']))
```

STEP 02: VISUALIZING DATA

```python
sns.pairplot(df_cancer, hue='target', vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'])
```

```python
sns.countplot(x=df_cancer['target'])
```

STEP 03: MODEL TRAINING 

```python
x = df_cancer.iloc[:,:-1]
# x = df_cancer.drop(['target'], axis = 1) 
```

```python
y = df_cancer.iloc[:,-1]
# y = df_cancer['target']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
svc_model.fit(x_train, y_train)
```

STEP 04: EVALUATE MODEL

```python
y_predict = svc_model.predict(x_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True)
```

STEP 05: IMPROVING MODEL

```python
min_train = x_train.min()
```

```python
range_train = (x_train-min_train).max()
```

```python
x_train_scaled = (x_train - min_train)/range_train
```

```python
sns.scatterplot(x = x_train['mean area'], y = x_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = x_train_scaled['mean area'], y = x_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = x_test.min()
```

```python
range_test = (x_test - min_test).max()
```

```python
x_test_scaled = (x_test - min_test)/range_test
```

```python
svc_model.fit(x_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(x_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
cm
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test,y_predict))
```

IMPROVING THE MODEL PART 02

```python
param_grid = {'C' : [0.1, 1, 100, 100], 'gamma' : [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(x_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(x_test_scaled)
```

```python
grid_cm = confusion_matrix(y_test, grid_predictions)
```

```python
grid_cm
```

```python
sns.heatmap(grid_cm, annot = True)
```

```python
print(classification_report(y_test, grid_predictions))
```
